```python
##Create Helper Functions for visualizations
def printMatrix(arr):
    string = ""
    for i in range(1,65):
        string += translation(arr[i-1]) + ' '
        if (i%8 == 0):
            print(string)
            string= ""
def translation(intval):
    if intval == 0:
        return '_'
    elif intval == 1:
        return 'R'
    elif intval == 2:
        return 'K'
    elif intval == 3:
        return 'B'
    elif intval ==4:
        return 'Q'
    elif intval == 5:
        return 'K'
    elif intval == 6:
        return 'B'
    elif intval == 9:
        return 'C'
    elif intval == 10:
        return 'N'
    elif intval == 11:
        return 'R'
    elif intval == 12:
        return 'K'
    elif intval == 13:
        return 'B'
    elif intval == 14:
        return 'Q'
    elif intval == 15:
        return 'K'
    elif intval == 16:
        return 'B'
```

```python
#Create 'Attack Functions'
def rook(index, weight = 1):
    baseline = np.zeros((8,8), dtype=np.int)
    x = index[0]
    y = index[1]
    for i in range(0,8):
        try:
            baseline[x,i] = weight
        except:
            pass
    for i in range(0,8):
        try:
            baseline[i,y] = weight
        except:
            pass
    baseline[x,y] = 0
    
    return baseline
            
```

```python
def bishop(index, weight = 1):
    baseline = np.zeros((8,8), dtype=np.int)
    x = index[0]
    y = index[1]
    #upper left
    for i in range(0,8):
        try:
            x1 = x - i
            y1 = y - i
            if (x1>=0) and (y1>=0): 
                baseline[x1, y1] = weight
                #print(baseline)
        except:
            pass
    #upper right    
    for i in range(0,8):
        try:
            x1 = x - i
            y1 = y + i
            if (x1>=0) and (y1>=0): 
                baseline[x1, y1] = weight
                #print(baseline)
        except:
            pass
    #lower right
    for i in range(0,8):
        try:
            x1 = x + i
            y1 = y + i
            if (x1>=0) and (y1>=0): 
                baseline[x1, y1] = weight
                #print(baseline)
        except:
            pass
        
    for i in range(0,8):
        try:
            x1 = x + i
            y1 = y - i
            if (x1>=0) and (y1>=0): 
                baseline[x1, y1] = weight
                #print(baseline)
        except:
            pass

    
    baseline[x,y] = 0
    return baseline
```

```python
def queen(index, weight = 1):
    baseline = bishop(index)
    baseline += rook(index)
    return baseline
```

```python
def king(index, weight = 1):
    baseline = np.zeros((8,8), dtype=np.int)
    x = index[0]
    y = index[1]
    try:
        x1 = x +1
        y1 = y
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    try:
        x1 = x +1
        y1 = y +1
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    try:
        x1 = x
        y1 = y +1
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    try:
        x1 = x +1
        y1 = y +1
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    try:
        x1 = x -1
        y1 = y -1
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    try:
        x1 = x
        y1 = y -1
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    try:
        x1 = x - 1
        y1 = y
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    try:
        x1 = x - 1
        y1 = y + 1
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    try:
        x1 = x + 1
        y1 = y - 1
        if (x1>=0) and (y1>=0): 
            baseline[x1, y1] = weight
    except:
        pass
    baseline[x,y] = 0
    return baseline
```
